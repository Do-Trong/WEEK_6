# Interpolation

Hai kĩ thật đơn giản


```python
import cv2
from matplotlib import pyplot as plt
img = cv2.imread("image_2.jpg",0)
(height, width) = img.shape
dim  = ( width*2, height*2)
resize_nearest = cv2.resize(img, dim, interpolation = cv2.INTER_NEAREST)
resize_linear = cv2.resize(img, dim, interpolation = cv2.INTER_LINEAR)
cv2.imshow("image",img)
cv2.imshow("resize_nearest",resize_nearest)
cv2.imshow("resize_linear",resize_linear)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
